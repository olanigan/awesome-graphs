```python
import gradio as gr  
from gradio_folium import Folium  
import folium  
import pandas as pd  
from pathlib import Path  

def load_data():  
    # Load and prepare the data  
    file_path = Path(r"C:\Users\pc\Downloads\dataset folder\Border_Crossing_Entry_Data.csv")  
    print(f"Loading data from: {file_path}")  
    df = pd.read_csv(file_path)  

    # Clean the coordinate data  
    df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')  
    df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')  
    df = df.dropna(subset=['Latitude', 'Longitude'])  

    print(f"Loaded data shape: {df.shape}")  
    print("\nFirst few rows of coordinates:")  
    print(df[['Latitude', 'Longitude']].head())  

    return df  

def select(evt: gr.SelectData):  
    row = df.iloc[evt.index[0]]  
    m = folium.Map(location=[row['Latitude'], row['Longitude']], zoom_start=10)  
    folium.Marker(  
        [row['Latitude'], row['Longitude']],  
        popup=f"Location: {row['Port Name']}" if 'Port Name' in row else "Selected Location"  
    ).add_to(m)  
    return m  

try:  
    print("Loading data...")  
    df = load_data()  

    print("\nCreating Gradio interface...")  
    with gr.Blocks() as demo:  
        gr.Markdown("# Border Crossing Entry Data Visualization")  

        initial_location = [df['Latitude'].iloc[0], df['Longitude'].iloc[0]]  
        initial_map = folium.Map(location=initial_location, zoom_start=4)  
        folium.Marker(initial_location, popup="Initial Location").add_to(initial_map)  

        with gr.Row():  
            dataframe = gr.Dataframe(  
                value=df,  
                headers=list(df.columns),  
                row_count=(10, "dynamic"),  
                interactive=False,  
                wrap=True  
            )  

            map_component = Folium(value=initial_map, height=500)  

        dataframe.select(fn=select, outputs=map_component)  

    print("\nLaunching application...")  
    demo.launch(share=True)  

except Exception as e:  
    print(f"\nError occurred: {str(e)}")  
```
